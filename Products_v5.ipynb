```python
# maps

# product map
def product_map(id):
    if id == 1046 or id == 1025:
        return 'play1'
    elif id == 1021 or id == 1020:
        return 'play3'
    elif id == 1016 or id == 1012:
        return 'play5'
    elif id == 1006:
        return 'connect'
    elif id == 1007:
        return 'connectAMP'
    elif id == 1023 or id == 1027 or id == 1028 or id == 1029:
        return 'sub'
    elif id == 1008:
        return 'bridge'
    elif id == 1043:
        return 'boost'
    else:
        return 'playbar'
    
def channel_map(c):
    if c == 'DIRECT' or c == 'NA Direct' or c == 'Direct':
        return 'direct'
    elif c == 'MASS RETAILER' or c == 'Mass Retailer':
        return 'mass'
    elif c == 'AUDIO VISUAL' or c == 'Audio Visual':
        return 'av'
    elif c == 'DISTRIBUTOR' or c == 'Distributor':
        return 'dist'
    else:
        return 'other'
```
