```python
indexer = agg_counts.sum(1).argsort()
indexer[:10]
```

```python
count_subset = agg_counts.take(indexer)[-10:]
count_subset
```

```python
normed_subset = count_subset.div(count_subset.sum(1), axis = 0)
normed_subset.plot(kind='barh', stacked = True)
```

```python
unames = ['user_id','gender','age','occupation','zip']
users = pd.read_table('pandas/ch02/movielens/users.dat',sep='::',header=None,names=unames)

rnames = ['user_id','movie_id','rating','timestamp']
ratings = pd.read_table('pandas/ch02/movielens/ratings.dat',sep='::',header=None,names=rnames)

mnames = ['movie_id','title','genres']
movies = pd.read_table('pandas/ch02/movielens/movies.dat',sep='::',header=None,names=mnames)
```

```python
#users.head()
#ratings.head()
#movies.head()
data = pd.merge(pd.merge(ratings,users),movies)
```

```python
data.head()
```

```python
data.ix[0]
```

```python
mean_ratings = data.pivot_table('rating',rows='title',cols='gender',aggfunc='mean')
mean_ratings[:5]
```

```python
ratings_by_title = data.groupby('title').size()
ratings_by_title[:15]
```

```python
active_titles = ratings_by_title.index[ratings_by_title>=250]
active_titles
```

```python
mean_ratings = mean_ratings.ix[active_titles]
mean_ratings
```

```python
top_female_ratings = mean_ratings.sort_index(by='F',ascending=False)
top_female_ratings[:10]
```

```python
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
```

```python
sorted_by_diff = mean_ratings.sort_index(by = 'diff', ascending= True)
sorted_by_diff[:10]
```

```python
sorted_by_diff[::-1][:10]
```

```python
rating_std_by_title = data.groupby('title')['rating'].std()
rating_std_by_title = rating_std_by_title.ix[active_titles]
rating_std_by_title.order(ascending=False)[:10]
```

```python
years = range(1880,2011)
pieces = []
columns = ['name','sex','births']
for year in years:
    path = 'pandas/ch02/names/yob%d.txt' % year
    df = pd.read_csv(path, names=columns)
    df['year'] = year
    pieces.append(df)
    names = pd.concat(pieces, ignore_index = True)
    
names
```

```python
total_births = names.pivot_table('births',rows='year',cols='sex',aggfunc=sum)
total_births
```

```python
total_births.plot(title='Total births by sex and year')
```

```python
def add_prop(group):
    births = group.births.astype(float)
    group['prop'] = births / births.sum()
    return group

names = names.groupby(['year','sex']).apply(add_prop)
```
